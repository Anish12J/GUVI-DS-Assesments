**Step 1: Importing the libraries**

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

**Step 2: Importing dataset**

```python
data = pd.read_csv("Data.csv")
data
```

**Step 3: Handling the missing data**

```python
data.isna().sum()
```

```python
data.fillna({'Salary':round(data["Salary"].mean(),1),'Age':round(data["Age"].mean(),1)},inplace = True)
data
```

**Step 4: Encoding categorical data**

```python
encoder = OneHotEncoder(dtype=np.int32)
dfEncodedFeatures = encoder.fit_transform(np.array(data["Country"]).reshape(-1,1))
columnsNames = encoder.categories_
df = pd.DataFrame(dfEncodedFeatures.toarray(),columns = columnsNames)
array = dfEncodedFeatures.toarray()
columnsNames
```

**Step 5: Creating a dummy variable**

```python
data.drop("Country",axis =1 ,inplace = True)
#i = 0
#for name in columnsNames:
#    data.insert(loc=i,column = name , value = array[:,i])
#    i=i + 1
data.insert(loc = 0,column = "France",value = array[:,0])
data.insert(loc = 1,column = "Germany",value = array[:,1])
data.insert(loc = 2,column = "Spain",value = array[:,2])
data
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 5)
```

**Step 7: Feature Scaling**

```python
stdScaler = StandardScaler()
tr_Xtrain = stdScaler.fit_transform(X_train[["Age","Salary"]])
tr_Xtest = stdScaler.fit_transform(X_test[["Age","Salary"]])
```

```python
tr_Xtrain = np.concatenate((tr_Xtrain,X_train.iloc[:,:3]),axis = 1)
tr_Xtest = np.concatenate((tr_Xtest,X_test.iloc[:,:3]),axis = 1)
```
